```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
df = pd.read_csv("../data/mega_millions.csv", header=None)
df.head()
```

```python
df.columns = ["mega", "month", "date", "year", "num_1", "num_2", "num_3", "num_4","num_5","megaball", "megaplier"]
```

```python
df["time"] = pd.to_datetime((df.year*10000+df.month*100+df.date).apply(str),format='%Y%m%d')
```

```python
df = df[["time", "num_1", "num_2", "num_3", "num_4","num_5"]]
df
```

```python
df.set_index("time")
```

```python
df.num_1.plot()
```

```python
df = df[df["time"]>"2017-10-28"].reset_index(drop=True)
df = df.set_index("time")
df.head()
```

```python
df.plot()
```

```python
df.corr()
```

```python
df.to_csv("output.csv")
```
