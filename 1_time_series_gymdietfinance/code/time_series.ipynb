```python
import pandas as pd
import seaborn as sns
sns.set()
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/gym_diet_finance.csv', skiprows=1)

df.head()
```

```python
df.Month = pd.to_datetime(df.Month)
```

```python
df.set_index("Month", inplace=True)
df.columns = ["diet", "gym", "finance"]
df.head()
```

```python
df.plot()
```

```python
df[["diet"]].plot()
```

```python
df.info()
```

```python
diet = df [["diet"]]
diet.diff().plot()
```

```python
diet.rolling(12).mean().plot()
```

```python
pd.plotting.autocorrelation_plot(diet)
```

```python
df.corr()
```

```python
gym_diet = pd.concat([df.diet.rolling(12).mean(), df.gym.rolling(12).mean()], axis=1)
gym_diet.plot()
```
